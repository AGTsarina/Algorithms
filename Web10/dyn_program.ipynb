Наивная рекурсия - НЕЭФФЕКТИВНО

```python
def fib_naive(n: int) -> int:
    if n <= 1:
        return n
    return fib_naive(n-1) + fib_naive(n-2)
```

```python
%%time
fib_naive(20)
```

ДП с запоминанием

```python
def fib_dp(n, memo):
    if memo[n] > 0:
        return memo[n]
    if n <= 1:
        memo[n] = n
        return n
    memo[n] = fib_dp(n-1, memo) + fib_dp(n-2, memo)
    return memo[n]
```

```python
%%time
n = 20
memo = [0] * (n + 1)
fib_dp(n, memo)
```

ДП Восходящее программирование

```python
def fib_bu(n):
    if n <= 1:
        return n
    dp = [0] *(n + 1)
    dp[1] = 1
    for i in range(2, n + 1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]
```

```python
%%time
n = 20
fib_bu(n)
```

Оптимизация хранения

```python
def fib_mem_opt(n):
    if n <= 1:
        return n
    prev0, prev1 = 0, 1
    res = 0
    for i in range(3, n + 1):
        res = prev0 + prev1
        prev0 = prev1
        prev1 = res
    return res
```

```python
%%time
n = 20
fib_mem_opt(n)
```

Задача о кузнечике

```python
def grasshopper_jumps(n):
    if n <= 0: return 0
    if n == 1: return 1
    if n == 2: return 1
    if n == 3: return 2

    a, b, c, d = 1, 1, 2, 0
    for i in range(4, n + 1):
        d = a + b + c
        a, b, c = b, c, d
    return d

grasshopper_jumps(20)
```

Наибольшая возрастающая подпоследовательность

```python
%%time
def lis(nums):
    n = len(nums)
    dp = [1] * n
    maxLen = 1
    for i in range(1, n):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)
        maxLen = max(maxLen, dp[i])
    return maxLen, dp


nums = [10, 9, 2, 5, 3, 7, 101, 18]
#length, dp = lis(nums)
#print(f'Длина наибольшей возрастающей последовательности {length}')
#print(f'Восстановить последовательность можно по массиву dp {dp}')
lis(nums)
```

Оптимизация

```python
%%time
def binary_search(arr, x):
    def search(L, R):
        if L == R: return L
        m = (L + R) // 2
        if x <= arr[m]: return search(L, m)
        return search(m + 1, R)
    if not arr or arr[-1] < x:
        return None
    return search(0, len(arr) - 1)

def lis_opt(nums):
    if not nums:
        return 0
    tails = []

    for x in nums:
        # Бинарный поиск позиции для вставки
        ind = binary_search(tails, x)

        if ind is None:
            # x больше всех элементов - добавляем в конец
            tails.append(x)
        else:
            # Заменяем элемент на x
            tails[ind] = x
    return tails

nums = [1, 9, 2, 5, 3, 7, 101, 18]
lis_opt(nums)
```

Максимальная сумма подмассива
Алгоритм Кадане

```python
# Расширенная версия, которая возвращает и индексы
def maxSubarraySumWithIndices(arr):
    n = len(arr)
    if n == 0: return 0, 0, 0

    max_ending_here = arr[0]
    max_so_far = arr[0]
    start, end, temp_start = 0, 0, 0;

    for i in range(1, n):
        if arr[i] > max_ending_here + arr[i]:
            max_ending_here = arr[i]
            temp_start = i
        else:
            max_ending_here += arr[i]

        if max_ending_here > max_so_far:
            max_so_far = max_ending_here
            start = temp_start
            end = i
    return max_so_far, start, end
arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
s, first, last = maxSubarraySumWithIndices(arr)
print(f'Наибольшая сумма: {s} для подмассива {arr[first:last + 1]}' )
```

### Задача о рюкзаке
1. Базовая версия. Динамическая таблица

```python
import numpy as np
def knapsack(W, weights, values):
    n = len(weights)
    dp = np.zeros((n + 1, W + 1))
    for i in range(1, n + 1):
        for w in range(1, W + 1):
            if weights[i-1] <= w:
                dp[i][w] = max(dp[i-1][w],
                              values[i-1] + dp[i-1][w - weights[i-1]])
            else:
                dp[i][w] = dp[i-1][w]

    # Восстановление выбранных предметов
    selected = []
    w = W
    for i in range(n, 0, -1):
        if dp[i][w] != dp[i-1][w]:
            selected.append(i-1)
            w -= weights[i-1]
        if w < 0: break

    selected_weights = [weights[i] for i in selected]
    selected_values = [values[i] for i in selected]

    #print(f"Выбранные вещи: {selected} с весами {selected_weights} с суммарным весом {sum(selected_weights)}")
    print(f"Выбранные вещи: {selected} с весами {selected_weights} со стоимостью {selected_values}")
    return dp[n][W]

values = [60, 100, 120, 150, 110, 70, 200, 100]
weights = [10, 20, 30, 15, 25, 20, 40, 10]
W = 50

print(f"Максимальная стоимость: {knapsack(W, weights, values)}")
```

2. Оптимизация памяти

```python
def knapsack(W, weights, values):
    n = len(weights)
    dp = np.zeros(W + 1)

    for i in range(n):
        for w in range(W,  weights[i] - 1, -1):
            dp[w] = max(dp[w], values[i] + dp[w - weights[i]])
    return dp[W]
print(f"Максимальная стоимость: {knapsack(W, weights, values)}")
```

3. Метод ветвей и границ

```python
'''struct Item {
    int weight, value;
    double ratio;
};'''

class BranchBoundKnapsack:
    def __init__(self, W) -> None:
        self.items = []
        self.capacity = W
        self.best_value = 0

    def upperBound(self, level, current_weight, current_value):
        remaining_capacity = self.capacity - current_weight;
        bound = current_value
        i = level

        while i < len(self.items) and self.items[i]["weight"] <= remaining_capacity:
            remaining_capacity -= self.items[i]["weight"]
            bound += self.items[i]["value"]
            i += 1

        if i < len(self.items):
            bound += remaining_capacity * self.items[i]["ratio"]
        return bound


    def dfs(self, level, current_weight, current_value):
        if current_weight > self.capacity: return

        if level == len(self.items):
            self.best_value = max(self.best_value, current_value)
            return

        # Проверка верхней границы
        if self.upperBound(level, current_weight, current_value) <= self.best_value:
            return

        # Включить текущий предмет
        self.dfs(level + 1, current_weight + self.items[level]["weight"],
            current_value + self.items[level]["value"])

        # Исключить текущий предмет
        self.dfs(level + 1, current_weight, current_value)


    def solve(self, weights, values):
        self.best_value = 0

        self.items = [{"weight": weights[i], "value": values[i], "ratio": values[i] / weights[i]}
                      for i in range(len(weights))]

        # Сортируем предметы по убыванию отношения стоимость/вес
        self.items = sorted(self.items, key=lambda x: -x["ratio"])

        self.dfs(0, 0, 0)
        return self.best_value


values = [60, 100, 120, 150, 110, 70, 200, 100]
weights = [10, 20, 30, 15, 25, 20, 40, 10]
W = 50

bk = BranchBoundKnapsack(W)
bk.solve(weights, values)
```

4. Meet-in-the-Middle

```python
'''struct Solution {
    int weight, value;
};'''

def generateSubsets(W, weights, values, start, end):
    n = end - start
    solutions = []

    for mask in range(1 << n): # все возможные варианты
        total_weight = 0
        total_value = 0
        for i in range(n):
            if mask & (1 << i):  # если вещь есть в данном варианте сборки
                total_weight += weights[start + i]
                total_value += values[start + i]
        solutions.append({"w": total_weight, "v": total_value})


    solutions.sort(key=lambda x: (x['w'], -x['v']))

    filtered = []
    max_value = -1
    for sol in solutions:
        if sol['w'] <= W and sol['v'] > max_value:
            filtered.append(sol);
            max_value = sol['v']
    return filtered


def meetInMiddle(W, weights, values):
    n = len(weights)
    mid = n // 2

    # Генерируем подмножества для двух половин
    left = generateSubsets(W, weights, values, 0, mid);
    right = generateSubsets(W, weights, values, mid, n);

    # Сортируем правую часть по весу
    right.sort(key=lambda x: x['w'])

    # Префиксные максимумы для правой части
    prefix_max = [0] * len(right)
    for i in range(len(right)):
        prefix_max[i] = right[i]['v']
        if i > 0:
            prefix_max[i] = max(prefix_max[i], prefix_max[i-1])


    best_value = 0

    # Комбинируем решения
    for left_sol in left:
        if left_sol['w'] > W: continue

        remaining = W - left_sol['w']

        # Бинарный поиск в правой части
        l = 0
        r = len(right) - 1
        best_right = 0

        while l <= r:
            m = (l + r) // 2;
            if right[m]['w'] <= remaining:
                best_right = prefix_max[m]
                l = m + 1
            else:
                r = m - 1

        best_value = max(best_value, left_sol['v'] + best_right)
    return best_value


values = [60, 100, 120, 150, 110, 70, 200, 100]
weights = [10, 20, 30, 15, 25, 20, 40, 10]
W = 50

meetInMiddle(W, weights, values)
```
